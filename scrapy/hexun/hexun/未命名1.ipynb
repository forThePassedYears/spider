```python
import requests
import re

header = {'User-Agent': 'Mozilla/5.0 (X11; Linux x86_64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/68.0.3440.84 Safari/537.36'}
res = requests.get('http://blog.hexun.com/', headers=header)
```

```python
res
```

```python
res.text
```

```python
pat = 'href="(http://.*?.blog.hexun.com/.*?_d.html)"'
url = re.compile(pat,re.S).findall(res.text)
```

```python
url
```

```python
for i in url:
    print(i)
```

```python
for i in range(0,len(url)):
    print(i, url[i])
```
